```python
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

conn = sqlite3.connect("bluestock_mf.db")
```

```python
nav = pd.read_sql(
    "SELECT * FROM fact_nav",
    conn
)

txn = pd.read_sql(
    "SELECT * FROM fact_transactions",
    conn
)

funds = pd.read_sql(
    "SELECT * FROM dim_fund",
    conn
)

perf = pd.read_sql(
    "SELECT * FROM fact_performance",
    conn
)

portfolio = pd.read_csv(
    "data/raw/09_portfolio_holdings.csv"
)
```

```python
nav['date'] = pd.to_datetime(nav['date'])

txn['transaction_date'] = pd.to_datetime(
    txn['transaction_date']
)
```

# Historical VaR and CVaR

```python
nav = nav.sort_values(
    ['amfi_code','date']
)

nav['daily_return'] = (
    nav.groupby('amfi_code')['nav']
    .pct_change()
)
```

# Create VaR Report

```python
results = []

for fund in nav['amfi_code'].unique():

    temp = nav[
        nav['amfi_code']==fund
    ]

    returns = temp[
        'daily_return'
    ].dropna()

    var95 = np.percentile(
        returns,
        5
    )

    cvar95 = returns[
        returns<=var95
    ].mean()

    results.append(
        [fund,var95,cvar95]
    )

var_df = pd.DataFrame(
    results,
    columns=[
        'amfi_code',
        'VaR95',
        'CVaR95'
    ]
)

var_df.head()
```

```python
var_df.to_csv(
    "var_cvar_report.csv",
    index=False
)
```

# Rolling 90-Day Sharpe

```python
top5 = nav[
    'amfi_code'
].unique()[:5]
```

```python
plt.figure(figsize=(14,6))

for fund in top5:

    temp = nav[
        nav['amfi_code']==fund
    ]

    sharpe = (
        temp['daily_return']
        .rolling(90)
        .mean()
        /
        temp['daily_return']
        .rolling(90)
        .std()
    ) * np.sqrt(252)

    plt.plot(
        temp['date'],
        sharpe,
        label=str(fund)
    )

plt.legend()

plt.title(
"Rolling 90-Day Sharpe Ratio"
)

plt.show()
```

# Investor Cohort Analysis

```python
first_txn = txn.groupby(
    'investor_id'
)['transaction_date'].min()

txn['cohort_year'] = txn[
    'investor_id'
].map(
    first_txn.dt.year
)
```

```python
cohort = txn.groupby(
    'cohort_year'
).agg(
    avg_sip=('amount_inr','mean'),
    total_invested=('amount_inr','sum')
)

cohort
```

# SIP Continuity

```python
sip = txn[
    txn['transaction_type']=="SIP"
].copy()
```

```python
counts = sip.groupby(
    'investor_id'
).size()

eligible = counts[
    counts>=6
].index

sip = sip[
    sip['investor_id'].isin(
        eligible
    )
]
```

```python
risk_list=[]

for inv in sip[
    'investor_id'
].unique():

    temp=sip[
        sip['investor_id']==inv
    ].sort_values(
        'transaction_date'
    )

    gap=temp[
        'transaction_date'
    ].diff().dt.days.mean()

    status="At Risk" if gap>35 else "Healthy"

    risk_list.append(
        [inv,gap,status]
    )

risk_df=pd.DataFrame(
    risk_list,
    columns=[
        'investor_id',
        'avg_gap_days',
        'status'
    ]
)

risk_df.head()
```
