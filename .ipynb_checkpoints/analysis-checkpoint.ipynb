```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
conn = sqlite3.connect("bluestock_mf.db")
```

```python
query = """
SELECT name
FROM sqlite_master
WHERE type='table';
"""

pd.read_sql(query, conn)
```

```python
pd.read_sql("SELECT * FROM fact_transactions LIMIT 5;", conn)
```

```python
query = """
SELECT state,
SUM(amount_inr) AS total_investment
FROM fact_transactions
GROUP BY state
ORDER BY total_investment DESC
LIMIT 10;
"""

state_df = pd.read_sql(query, conn)

state_df
```

```python
plt.figure(figsize=(10,5))
plt.bar(state_df['state'], state_df['total_investment'])
plt.xticks(rotation=45)
plt.title("Top 10 States by Investment")
plt.show()
```

```python
query = """
SELECT gender,
SUM(amount_inr) AS total_investment
FROM fact_transactions
GROUP BY gender;
"""

gender_df = pd.read_sql(query, conn)

gender_df
```

```python
plt.figure(figsize=(5,5))
plt.pie(
    gender_df['total_investment'],
    labels=gender_df['gender'],
    autopct='%1.1f%%'
)
plt.title("Gender Wise Investment")
plt.show()
```

```python
query = """
SELECT age_group,
SUM(amount_inr) AS total_investment
FROM fact_transactions
GROUP BY age_group
ORDER BY total_investment DESC;
"""

age_df = pd.read_sql(query, conn)

age_df
```

```python
plt.figure(figsize=(8,5))
plt.bar(age_df['age_group'], age_df['total_investment'])
plt.title("Age Group Investment")
plt.show()
```

```python
query = """
SELECT transaction_type,
COUNT(*) AS total_transactions
FROM fact_transactions
GROUP BY transaction_type;
"""

txn_df = pd.read_sql(query, conn)

txn_df
```

```python
plt.figure(figsize=(6,5))
plt.bar(txn_df['transaction_type'], txn_df['total_transactions'])
plt.title("Transaction Types")
plt.show()
```

```python
query = """
SELECT scheme_name,
return_5yr_pct
FROM fact_performance
ORDER BY return_5yr_pct DESC
LIMIT 10;
"""

top_funds = pd.read_sql(query, conn)

top_funds
```

```python
query = """
SELECT fund_house,
MAX(aum_crore) AS total_aum
FROM fact_aum
GROUP BY fund_house
ORDER BY total_aum DESC
LIMIT 10;
"""

aum_df = pd.read_sql(query, conn)

aum_df
```

```python
plt.figure(figsize=(10,5))
plt.bar(aum_df['fund_house'], aum_df['total_aum'])
plt.xticks(rotation=90)
plt.title("Top Fund Houses by AUM")
plt.show()
```

1. State Punjab has the highest investment amount.
2. Age Group 26-35 invests the most.
3. SIP is the most preferred transaction type.
4. Fund ABSL Small Cap Fund - Regular - Growth gives the highest 5-year return.
5. Fund House SBI Mutual Fund manages the largest AUM.
6. Male/Female investors contribute more investment.
